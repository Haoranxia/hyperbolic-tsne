```python
import os
from pathlib import Path 
import traceback
import numpy as np
from numpy import linalg as LA
from matplotlib import pyplot as plt
import json
```

```python
def plot_embeddings(emb, labels=None):
    fig, ax = plt.subplots()
    ax.scatter(emb[:, 0], emb[:, 1],
               c=labels,
               marker='.',
               s=1,             # size of points
               cmap="tab10")
    
    ax.add_patch(plt.Circle((0, 0), radius=1, edgecolor="b", facecolor="None"))
    ax.set_aspect('equal', adjustable='box')
    ax.axis("off")
    
    return fig
```

```python
save_loc = "experiments_thesis/exp_5.3/"
```

```python
## C_ELEGANS 
# lr = 6.4 (32 * 0.2)
dataset = "C_ELEGANS"
label_path = f"experiments_thesis/GaussianKL/{dataset}/experiment_4/labels.csv"
labels = np.loadtxt(label_path)

correct_grad_path = f"experiments_thesis/GaussianKL/{dataset}/experiment_4"
correct_emb = np.loadtxt(correct_grad_path + "/emb.csv", delimiter=',')
fig_2 = plot_embeddings(correct_emb, labels=labels)
fig_2.savefig(
    save_loc + f"{dataset}_gauss_lr6_4_5k.pdf",  # Filename
    dpi=300,          # Resolution in dots per inch (default is 100)
    bbox_inches="tight",  # Adjusts the bounding box to fit the content
    transparent=True  # Saves the figure with a transparent background
)

plt.show()
```
